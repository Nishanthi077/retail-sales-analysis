```python
import pandas as pd

# Load your dataset
data = pd.read_csv("retail_sales_dataset.csv")

# Show first 5 rows
data.head()
```

```python
data.columns
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
sales_over_time = data.groupby('Date')['Total Amount'].sum().reset_index()
```

```python
plt.figure(figsize=(12,6))
plt.plot(sales_over_time['Date'], sales_over_time['Total Amount'], marker='o')
plt.title('Total Sales Over Time', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```

```python

data['Month'] = data['Date'].dt.to_period('M')


monthly_sales = data.groupby('Month')['Total Amount'].sum().reset_index()


plt.figure(figsize=(12,6))
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Total Amount'], marker='o', color='orange')
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt

top_categories = data.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_categories.plot(kind='bar', color='skyblue')
plt.title('Top 10 Product Categories by Total Sales', fontsize=14)
plt.xlabel('Product Category')
plt.ylabel('Total Sales (â‚¹)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_customers = data.groupby('Customer ID')['Total Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_customers.plot(kind='bar', color='salmon')
plt.title('Top 10 Customers by Total Spending', fontsize=14)
plt.xlabel('Customer ID')
plt.ylabel('Total Spending (â‚¹)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
top_quantity = data.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_quantity.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Product Categories by Quantity Sold', fontsize=14)
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(x='Price per Unit', y='Quantity', data=data, color='orange')
plt.title('Relationship between Price per Unit and Quantity Sold')
plt.xlabel('Price per Unit (â‚¹)')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
avg_spending_gender = data.groupby('Gender')['Total Amount'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,5))
avg_spending_gender.plot(kind='bar', color=['#ff9999','#66b3ff'])
plt.title('Average Spending by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Total Amount (â‚¹)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


numeric_data = data[['Age', 'Quantity', 'Price per Unit', 'Total Amount']]


plt.figure(figsize=(8,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features', fontsize=14)
plt.show()
```
